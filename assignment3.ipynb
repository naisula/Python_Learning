1. Write a program that generates a random number between 1 and 100. The user has to guess the number in a limited number of attempts. Provide feedback after each guess (higher, lower, or correct).

```python
import random

def guessing_game(): # defined a function called guessing game
   rand_number = random.randint(1,100) # a random number is picked between 1 and 100
   attempts = 5 
   print("Guess a number between 1 and 100")

   for numbers in range(attempts): # runs the block 5 times since attempts = 5 
     guess = int(input("Enter your guess: ")) # the user is re quired to enter input as integer

     # if the users guess = random number generated the loop stops here and returnsa message 
     if guess == rand_number:
       return "Correct! You guessed the right number"
     elif guess < rand_number:
       print ("Too low !")
     else:
       print ("Too High !")
   return f"Game over ! The number was {rand_number}"# this message is displayed if the user attempts more than the given attempts

print(guessing_game()) #you have  to call the function to get an output on the screen

```

2. Write a program that allows the user to enter two numbers and an operator (+, -, *, /) and perform the corresponding calculation. Use conditional statements to handle different operators.

```python
#promts the user to enter 2 numbers and store them as floats
num1 = float(input("Enter first number:"))
num2 = float(input("Enter second number:"))
#prompts the user to enter an operand
operand = input("Enter operator(+,-,/,*): ")
if operand == "+": # if + :addition
    result = num1 + num2
elif operand == "-": # if -:subtraction
    result = num1 - num2
elif operand == "/": # if /:division
    result = num1 / num2
elif operand == "*": #if *: multiplication
    result = num1 * num2
else :
    result = "Invalid operand" 
print(result)




```

3. Write a program that uses a while loop to check if a number entered by the user is a prime number (divisible only by 1 and itself).

```python
prime_num = int(input("Enter Number: "))# asks the user for input and stores the variable as an integer. Prime numbers have to be whole numbers
i = 2 # starts from two because 1 is not prime
while i < prime_num: # the loop runs while i < the users input
 if prime_num % i == 0: # this checks for divisibility
  print("Number is not prime") # if the remainder is zero the number is not prime
  break # exits the loop then incremenys i and checks if prime
 i += 1
else: # if prime loop finishes and we get an output
    print("Number is prime")
```

4. Write a program that uses a for loop to print the multiplication table of a number entered by the user (up to a specified number of multiples).

```python
def multiplication_table(n): # creates a function that wil use a number n
    for i in range(1,10):# the variable i takes the next number each time the loop runs (1-10)
        print(f"{n} * {i} = {n*i}")
       
num = int(input("Enter Number: ")) # prompts the user for an input and stores the variable as an integer
multiplication_table(num)
```

5. Write a program that uses a for loop to calculate the sum of numbers from 1 to a given number entered by the user.

```python
num = int(input("Enter Last Number: ")) 
total = 0 # this variable stores the running total. sum starts at zero
for i in range(1, num + 1):# loops from 1 to the number the user enters
    total = total + i # add each number to the total
print(f"The sum from 1 to {num} is {total}")



```

6. Write a program that uses a for loop to print a pattern of stars or characters like a triangle.

```python
rows = int(input("Enter number of rows: "))
for i in range(1, rows + 1):
    print("*" * i) # "*" is a string and i repeats it i times
```
